BAR PLOT IN SEABORN LIBRARY:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
df1["island"]
```

```python
sns.barplot(x = df1.island, y = df1.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data =df1)
plt.show()
```

```python
df1
```

```python
sns.barplot(x ="island", y = 'bill_length_mm', data = df1, hue = "sex")
plt.show()
```

ORDER:

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", order = order1)
plt.show()
```

HUE ORDER:

```python
order1 = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df1, hue = "sex", 
            order = order1, hue_order = ['Female', 'Male'] )
plt.show()
```

```python
order1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, color = "pink")
plt.show()
```

```python
order1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = "sex", palette = "Accent")
plt.show()
```

```python
order1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = "sex", palette = "gist_earth")
plt.show()
```

SATURATION:

```python
order1 = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = "sex", palette = "gist_earth", saturation = 0.3)
plt.show()
```

HISTOGRAM IN SEABORN LIBRARY:

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1
```

```python
sns.displot(df1["flipper_length_mm"])
plt.show()
```

BINS:

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230])
plt.show()
```

COLOR:

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230], color = "y")
plt.show()
```

KDE:

```python
sns.displot(df1["flipper_length_mm"], bins = [170,180,190,200,210,220,230],kde = True, color = "blue")
plt.show()
```
